```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Movies_Path = "output/Final_Data.csv"
Movie_DF = pd.read_csv(Movies_Path)
```

```python
Movie_DF
```

```python
Movie_DF["ROI"] =(Movie_DF["Domestic_Gross"]/Movie_DF["Budget"]) *100
Movie_DF["ROI"] = Movie_DF["ROI"].round()
#Run_Groups = Movie_DF.groupby("Runtime_Min")
#Run_Counter = Run_Groups["Title"].nunique()

#Run_Grouping = pd.DataFrame({"Total Count": age_Counter, "Percent of Players": age_Per})
#age_Grouping["Percent of Players"] = age_Grouping["Percent of Players"].map("{:.2f}%".format)
#age_Grouping
Run_Grouping = pd.DataFrame(Movie_DF)
Run_Grouping= Run_Grouping.dropna()
Run_Grouping["ROI"] = Run_Grouping["ROI"].astype("float")
Run_Grouping["Runtime_Min"] = Run_Grouping["Runtime_Min"].astype("int")
Run_Grouping
```

```python
df_filtered = Run_Grouping[Run_Grouping['Runtime_Min'] >= 74] 
df_filtered = df_filtered[df_filtered['Runtime_Min'] <= 105] 
# Delete these row indexes from dataFrame
df_filtered
```

```python
df_filtered["Bud_Per_Minute"] = df_filtered["Budget"]/df_filtered["Runtime_Min"]
df_filtered["Bud_Per_Minute"] = df_filtered["Bud_Per_Minute"].round()
df_filtered
```

```python
Run_Plot = df_filtered.groupby("Runtime_Min")
histo = Run_Plot.apply(lambda x: x.mean())

#Run_Plot.apply()
#Run_Plot.reset_index(drop=True, inplace=True)
#Run_Plot = Run_Grouping.groupby( [ "RunTime"]).count().reset_index()
#test = Run_Plot
#test = test.reset_index
#Run_Plot.aggregate(np.mean)

#est_DF = pd.DataFrame("Run_Plot")
#Run_Plot["ROI"]=Run_Plot["ROI"].astype("float")
#df= pd.DataFrame("Run_Plot")
#histo.reset_index(level=0, inplace=True)
histo
#histo = histo.reset_index

histo1 = pd.DataFrame(histo)

histo1['index_col'] = histo1.index
histo1.reset_index
histo1.set_index("index_col")
```

```python
#fig, run_plt = plt.plot(histo["RunTime"],histo["ROI"])



plt.figure(figsize=(15,15))
plt.plot(histo["Runtime_Min"],histo["Bud_Per_Minute"])
plt.xlabel("RunTime (Min.)")
plt.ylabel("Budget Per Minute")
plt.title("Average Budget Per Minute V Runtime")
plt.savefig("Output/Budget V Run.png")
```
