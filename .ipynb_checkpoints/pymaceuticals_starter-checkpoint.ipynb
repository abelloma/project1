```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem, ttest_ind

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouse_drug = pd.read_csv(mouse_drug_data_to_load)
clinical_trial = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
drug_clinical = pd.merge(clinical_trial, mouse_drug, on="Mouse ID", how="inner")


# Display the data table for preview
drug_clinical.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_gb = drug_clinical.groupby(["Drug", "Timepoint"])['Tumor Volume (mm3)'].mean()
# print (drug_gb)

# tumor = drug_clinical['Tumor Volume (mm3)'].mean()

# Convert to DataFrame

drug_df = pd.DataFrame(drug_gb).reset_index()
# tumor_df = pd.DataFrame(tumor)
# # Preview DataFrame
drug_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# drug_df.sem(axis=1)
# Convert to DataFrame

standard_error_gb = drug_clinical.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()
standard_error_df = pd.DataFrame(standard_error_gb).reset_index()

standard_error_df.head()

# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames
timepoint = drug_df.set_index(['Timepoint','Drug']).unstack('Drug')
# timepoint = drug_df.transpose().unstack('Drug')

# Preview that Reformatting worked
timepoint.head()
```

```python
#make dataframe with mean and sem

mean_sem = pd.merge(drug_df, standard_error_df, on=("Drug","Timepoint"),how="inner")
mean_sem_df = mean_sem.rename(columns={"Tumor Volume (mm3)_x" : "mean", "Tumor Volume (mm3)_y" : "sem"})
mean_sem_df.head()
```

```python
#get means and sem in list by drug
cap_tvmean_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Capomulin', 'mean'].tolist()
inf_tvmean_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Infubinol', 'mean'].tolist()
ket_tvmean_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Ketapril', 'mean'].tolist()
pla_tvmean_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Placebo', 'mean'].tolist()

cap_tvsem_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Capomulin', 'sem'].tolist()
inf_tvsem_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Infubinol', 'sem'].tolist()
ket_tvsem_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Ketapril', 'sem'].tolist()
pla_tvsem_list = mean_sem_df.loc[mean_sem_df['Drug'] =='Placebo', 'sem'].tolist()
```

```python
# Generate the Plot (with Error Bars)

ax = plt.subplot(111)

# Set the x axis from 0 to 45 in increments of 5
x_axis = np.arange(0, 50, 5)

# Set the plot title and axes titles
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")

# Plot the 'mean' list vs. the established x axis with error 
ax.errorbar(x_axis, cap_tvmean_list, yerr=cap_tvsem_list, fmt="red",
                   label="Capomulin", marker='o')
ax.errorbar(x_axis, inf_tvmean_list, yerr=inf_tvsem_list, fmt="green",
                   label="Infubinol",marker='s')
ax.errorbar(x_axis, ket_tvmean_list, yerr=ket_tvsem_list, fmt="blue",
                   label="Ketapril",marker='x')
ax.errorbar(x_axis, pla_tvmean_list, yerr=pla_tvsem_list, fmt="black",
                   label="Placebo",marker='+')

# Add the legend and gridlines
ax.legend(loc=2)

tick_locations = [value for value in x_axis]
ax.set_xticks(tick_locations, minor=False)
ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.xlim(0, max(x_axis)+2)
# Save the Figure

plt.savefig("Images/Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
std_error_gb = drug_clinical.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
std_error_df = pd.DataFrame(std_error_gb).reset_index()
# Preview DataFrame
std_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic = meta_df.set_index(['Timepoint','Drug']).unstack('Drug')
# Preview that Reformatting worked
metastatic.head()
```

```python
#get means and sem in list by drug
cap_mrmean_list = meta_df.loc[meta_df['Drug'] =='Capomulin', 'Metastatic Sites'].tolist()
inf_mrmean_list = meta_df.loc[meta_df['Drug'] =='Infubinol', 'Metastatic Sites'].tolist()
ket_mrmean_list = meta_df.loc[meta_df['Drug'] =='Ketapril', 'Metastatic Sites'].tolist()
pla_mrmean_list = meta_df.loc[meta_df['Drug'] =='Placebo', 'Metastatic Sites'].tolist()

cap_mrsem_list = std_error_df.loc[std_error_df['Drug'] =='Capomulin', 'Metastatic Sites'].tolist()
inf_mrsem_list = std_error_df.loc[std_error_df['Drug'] =='Infubinol', 'Metastatic Sites'].tolist()
ket_mrsem_list = std_error_df.loc[std_error_df['Drug'] =='Ketapril', 'Metastatic Sites'].tolist()
pla_mrsem_list = std_error_df.loc[std_error_df['Drug'] =='Placebo', 'Metastatic Sites'].tolist()
```

```python
# Generate the Plot (with Error Bars)

ax = plt.subplot(111)

# Set the x axis from 0 to 45 in increments of 5
x_axis = np.arange(0, 50, 5)

# Set the plot title and axes titles
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (days)")
plt.ylabel("Met Sites")

# Plot the 'mean' list vs. the established x axis with error 
ax.errorbar(x_axis, cap_mrmean_list, yerr=cap_mrsem_list, fmt="red",
                   label="Capomulin", marker='o')
ax.errorbar(x_axis, inf_mrmean_list, yerr=inf_mrsem_list, fmt="green",
                   label="Infubinol",marker='s')
ax.errorbar(x_axis, ket_mrmean_list, yerr=ket_mrsem_list, fmt="blue",
                   label="Ketapril",marker='x')
ax.errorbar(x_axis, pla_mrmean_list, yerr=pla_mrsem_list, fmt="black",
                   label="Placebo", marker='+')

# Add the legend and gridlines
ax.legend(loc=2)

tick_locations = [value for value in x_axis]
ax.set_xticks(tick_locations, minor=False)
ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.xlim(0, max(x_axis)+2)

# Show the Figure
plt.savefig("Images/Spread.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
change = (timepoint.iloc[-1] - timepoint.iloc[0]) / timepoint.iloc[0] * 100

# Display the data to confirm
change
```

```python
# Store all Relevant Percent Changes into a Tuple

cap_tv_change = int(((cap_tvmean_list[-1] - cap_tvmean_list[0])/cap_tvmean_list[0])*100)
inf_tv_change = int(((inf_tvmean_list[-1] - inf_tvmean_list[0])/inf_tvmean_list[0])*100)
ket_tv_change = int(((ket_tvmean_list[-1] - ket_tvmean_list[0])/ket_tvmean_list[0])*100)
pla_tv_change = int(((pla_tvmean_list[-1] - pla_tvmean_list[0])/pla_tvmean_list[0])*100)

# change_list[]

# for drug in change:
#     if Drug == Capomulin:
#         change_list.append()
# print(change_list)

# y values using a list
y = [cap_tv_change, inf_tv_change, ket_tv_change, pla_tv_change]
# Set the x axis from 0 to 45 in increments of 5
x_axis = np.arange(len(y))
    
# Splice the data between passing and failing drugs
colors = []
for item in y:
    if item < 0:
        colors.append('green')
    else:
        colors.append('red')

# Orient widths. Add labels, tick marks, etc. 

growth_bar = plt.bar(x_axis, y, color=colors, alpha=1.0, edgecolor='black', linewidth=0.7)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

plt.hlines(0, -1, len(x_axis), alpha=1.0, linewidth=0.5)
plt.grid('on', which='major', axis='y', linestyle='dotted', linewidth=0.5)

# Use functions to label the percentages of changes
def autolabel(rects):
    for i,rect in enumerate(rects):
        height = int(y[i])
        if height >= 0:
            plt.text(rect.get_x()+rect.get_width()/2., 2, '%s%%'% (y[i]),
                ha='center', va='bottom', color='white', weight='bold')
        else:
            plt.text(rect.get_x()+rect.get_width()/2., -2, '%s%%'% (y[i]),
                ha='center', va='top', color='white', weight='bold')          

autolabel(growth_bar)

# Call functions to implement the function calls
autolabel(growth_bar)

# Save the Figure
plt.savefig("Images/CompareBar.png")

# Show the Figure
```
